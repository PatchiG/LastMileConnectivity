```python

import csv


# Function to get available sharing system networks
def get_available_networks(api_key):
    url = "https://external.transitapp.com/v3/map_layers/available_networks"

    # Set the header with the API Key
    headers = {
        'apiKey': api_key
    }

    # Make the API request with header
    response = requests.get(url, headers=headers)

    if response.status_code == 200:
        return response.json()  # Return JSON response
    else:
        return {"error": f"Failed to retrieve data: {response.status_code}"}

# Function to write networks to a CSV
def write_networks_to_csv(networks, filename='networks.csv'):
    # Define the CSV header
    header = ['network_id', 'network_name', 'network_color', 'coordinates']

    # Open the CSV file for writing
    with open(filename, mode='w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(header)  # Write header row

        # Loop through each network in the response
        for network in networks.get('networks', []):
            # Extract the necessary information
            network_id = network.get('network_id', '')
            network_name = network.get('network_name', '')
            network_color = network.get('network_color', '')
            coordinates = network.get('network_geometry', {}).get('geometry', {}).get('coordinates', [])

            # Flatten the coordinates for CSV
            coords_flat = str(coordinates)  # Convert list of coordinates to string

            # Write the network details to the CSV
            writer.writerow([network_id, network_name, network_color, coords_flat])

# Fetch available networks
networks = get_available_networks(api_key)

# Handle error if the API call fails
if "error" in networks:
    print(networks["error"])
else:
    # Write the networks data to a CSV file
    write_networks_to_csv(networks)
    print("Networks data has been written to networks.csv")
```

```python
import requests
import time
import csv
import json

# Define constants
API_URL = "https://external.transitapp.com/v3/map_layers/placemarks"
API_KEY = ""  # Replace with your actual API key
OUTPUT_JSON_FILE = "shared_mobility.json"
OUTPUT_CSV_FILE = "shared_mobility.csv"

# List of locations
locations = [
    ("Washington, DC", 38.9076, -77.0365),
    ("Bethesda, MD", 39.0040, -77.0890),
    ("Silver Spring, MD", 38.9907, -77.0261),
    ("Arlington, VA", 38.8870, -77.0941),
    ("Alexandria, VA", 38.8048, -77.0469),
    ("Rockville, MD", 39.0836, -77.1528),
    ("Gaithersburg, MD", 39.1434, -77.2013),
    ("Tysons, VA", 38.9277, -77.2210),
    ("Fairfax, VA", 38.8462, -77.3064),
    ("McLean, VA", 38.9331, -77.1793),
    ("Falls Church, VA", 38.8816, -77.1747),
    ("Vienna, VA", 38.9010, -77.2630),
    ("College Park, MD", 38.9867, -76.9378),
    ("Greenbelt, MD", 39.0028, -76.8756),
    ("Hyattsville, MD", 38.9680, -76.9480),
    ("Laurel, MD", 39.0997, -76.8483),
    ("Bowie, MD", 38.9837, -76.7792),
    ("Chantilly, VA", 38.8790, -77.4426),
    ("Reston, VA", 38.9587, -77.3570),
    ("Manassas, VA", 38.7509, -77.4753),
]

# Function to fetch data from the API
def fetch_vehicles(lat, lon, distance=1000):
    headers = {"apiKey": API_KEY}
    params = {"lat": lat, "lon": lon, "distance": distance}

    try:
        response = requests.get(API_URL, headers=headers, params=params)
        response.raise_for_status()  # Raise an exception for HTTP errors
        return response.json()  # Return JSON data as a Python dict
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data for ({lat}, {lon}): {e}")
        return None

# Save aggregated data to JSON
def save_to_json(data, filename):
    with open(filename, "w") as json_file:
        json.dump(data, json_file, indent=4)
    print(f"Aggregated data saved to {filename}")

# Save aggregated data to CSV
def save_to_csv(data, filename):
    if not data:
        print("No valid data to save in CSV. Exiting.")
        return

    with open(filename, mode="w", newline="", encoding="utf-8") as csv_file:
        fieldnames = data[0].keys()
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)
    print(f"Aggregated data saved to {filename}")

# Main script logic
def main():
    all_data = []  # List to store data for all locations
    for name, lat, lon in locations:
        print(f"Fetching data for {name}...")
        data = fetch_vehicles(lat, lon, distance=1000)
        if data and isinstance(data, dict) and "placemarks" in data:
            placemarks = data["placemarks"]
            # Add location name to each placemark
            for item in placemarks:
                item["location"] = name
            all_data.extend(placemarks)
        else:
            print(f"No valid data for {name}. Skipping.")
        time.sleep(60)  # Respect rate limits by pausing between requests

    # Save aggregated data to JSON and CSV
    save_to_json(all_data, OUTPUT_JSON_FILE)
    save_to_csv(all_data, OUTPUT_CSV_FILE)

    print("Data fetching for all locations completed.")

# Run the script
if __name__ == "__main__":
    main()
```

```python
import requests
import time
import csv
import json

# Define constants
API_URL = "https://external.transitapp.com/v3/map_layers/placemarks"
API_KEY = ""  # Replace with your actual API key
OUTPUT_JSON_FILE = "shared_mobility.json"
OUTPUT_CSV_FILE = "shared_mobility.csv"

# List of locations
locations = [
    ("Washington, DC", 38.9076, -77.0365),
    ("Bethesda, MD", 39.0040, -77.0890),
    ("Silver Spring, MD", 38.9907, -77.0261),
    ("Arlington, VA", 38.8870, -77.0941),
    ("Alexandria, VA", 38.8048, -77.0469),
    ("Rockville, MD", 39.0836, -77.1528),
    ("Gaithersburg, MD", 39.1434, -77.2013),
    ("Tysons, VA", 38.9277, -77.2210),
    ("Fairfax, VA", 38.8462, -77.3064),
    ("McLean, VA", 38.9331, -77.1793),
    ("Falls Church, VA", 38.8816, -77.1747),
    ("Vienna, VA", 38.9010, -77.2630),
    ("College Park, MD", 38.9867, -76.9378),
    ("Greenbelt, MD", 39.0028, -76.8756),
    ("Hyattsville, MD", 38.9680, -76.9480),
    ("Laurel, MD", 39.0997, -76.8483),
    ("Bowie, MD", 38.9837, -76.7792),
    ("Chantilly, VA", 38.8790, -77.4426),
    ("Reston, VA", 38.9587, -77.3570),
    ("Manassas, VA", 38.7509, -77.4753),
]

# Function to fetch data from the API
def fetch_vehicles(lat, lon, distance=1000):
    headers = {"apiKey": API_KEY}
    params = {"lat": lat, "lon": lon, "distance": distance}

    try:
        response = requests.get(API_URL, headers=headers, params=params)
        response.raise_for_status()  # Raise an exception for HTTP errors
        return response.json()  # Return JSON data as a Python dict
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data for ({lat}, {lon}): {e}")
        return None

# Save aggregated data to JSON
def save_to_json(data, filename):
    with open(filename, "w") as json_file:
        json.dump(data, json_file, indent=4)
    print(f"Aggregated data saved to {filename}")

# Save aggregated data to CSV
def save_to_csv(data, filename):
    if not data:
        print("No valid data to save in CSV. Exiting.")
        return

    with open(filename, mode="w", newline="", encoding="utf-8") as csv_file:
        fieldnames = data[0].keys()
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)
    print(f"Aggregated data saved to {filename}")

# Process placemarks to extract required fields and handle nested counts
def process_placemarks(placemarks, location_name):
    processed_data = []
    for item in placemarks:
        # Add location name
        item["location"] = location_name

        # Handle stationProperties['counts']
        counts = item.get("stationProperties", {}).get("counts", {})
        item["bike"] = counts.get("bike", 0)
        item["electric_bike"] = counts.get("electric-bike", 0)
        item["dock"] = counts.get("dock", 0)

        # Remove stationProperties if no longer needed
        item.pop("stationProperties", None)

        processed_data.append(item)
    return processed_data

# Main script logic
def main():
    all_data = []  # List to store data for all locations
    for name, lat, lon in locations:
        print(f"Fetching data for {name}...")
        data = fetch_vehicles(lat, lon, distance=1000)
        if data and isinstance(data, dict) and "placemarks" in data:
            placemarks = data["placemarks"]
            processed_data = process_placemarks(placemarks, name)
            all_data.extend(processed_data)
        else:
            print(f"No valid data for {name}. Skipping.")
        time.sleep(60)  # Respect rate limits by pausing between requests

    # Save aggregated data to JSON and CSV
    save_to_json(all_data, OUTPUT_JSON_FILE)
    save_to_csv(all_data, OUTPUT_CSV_FILE)

    print("Data fetching for all locations completed.")

# Run the script
if __name__ == "__main__":
    main()
```
